```python
import pandas as pd

# solicitar o generar las ventas de los últimos 5 días
ventas = []#estamos creando vector
for i in range(1, 6):#un blucle que dara 5
    venta_dia = float(input(f"Ingrese las ventas del día {i}: "))
    ventas.append(venta_dia)

# crear el DataFrame con los datos de las ventas
df_ventas = pd.DataFrame({'ventas': ventas}, index=['dia 1', 'dia 2', 'dia 3', 'dia 4', 'dia 5'])


# calcular el descuento del 5%
descuento = 0.05

# aplicar el descuento a las ventas y crear las nuevas columnas
df_ventas['ventas con descuento'] = df_ventas['ventas'] * (1 - descuento)
df_ventas['nro de unidades'] = [10, 20, 15, 30, 25] # ejemplo de cantidad de unidades por día
df_ventas['Prendas'] = ["Polo", "Short", "Boxer", "Falda", "Gorros"]
df_ventas['precio unitario'] = round(df_ventas['ventas'] / df_ventas['nro de unidades'],2)

# mostrar el DataFrame resultante
print(df_ventas)
```

```python
import pandas as pd

# Cargar el archivo csv en un DataFrame
df = pd.read_csv('datos_seguros.csv', sep=';')

# Mostrar las dimensiones del DataFrame
print("Dimensiones del DataFrame:")
print(df.shape)

# Mostrar los nombres de las columnas
print("\nNombres de las columnas:")
print(df.columns)

# Mostrar los tipos de datos de las columnas
print("\nTipos de datos de las columnas:")
print(df.dtypes)

# Mostrar las 10 primeras filas del DataFrame
print("\nLas 10 primeras filas:")
print(df.head(10))

# Mostrar las 10 últimas filas del DataFrame
print("\nLas 10 últimas filas:")
print(df.tail(10))

# Verificar si hay valores perdidos/nulos en el DataFrame
print("\n¿Hay valores perdidos/nulos?")
print(df.isnull().values.any())

# Extraer las primeras 1000 filas y almacenarlas en un contenedor
primeras_1000 = df[:1000]
print("\nLas primeras 1000 filas:")
print(primeras_1000)


# es una función de Pandas que se utiliza para establecer el número máximo de filas que se muestran en una salida de DataFrame. En este caso, None significa que no hay límite en el número de filas que se mostrarán y, por lo tanto, se mostrarán todas las filas del DataFrame en la salida
pd.set_option('display.max_rows', None)
primeras_100 = df[:100]
print("\nLas primeras 100 filas:")
print(primeras_100)
```
